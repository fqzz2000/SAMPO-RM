```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
cpovio = pd.read_csv('cpo_violation.csv',index_col=0)
cpor = pd.read_csv('cpo_mean_reward.csv',index_col=0)
newvio = pd.read_csv('newalgo_violation.csv',index_col=0)
newr = pd.read_csv('newalgo_mean_reward.csv',index_col=0)
```

```python
plt.figure(figsize=(10,7))
plt.plot(cpor,label='CPO')
plt.plot(newr,label='SAPO-RM')
plt.legend(fontsize=25)
plt.xlabel('The Number of Epoch',fontsize=20)
plt.ylabel('Average Return',fontsize=20)
plt.grid()
```

```python
plt.figure(figsize=(10,7))
plt.plot(cpovio[:300],label = 'CPO')
plt.plot(newvio[:300],label = 'SAPO-RM')
plt.legend(fontsize=25)
plt.xlabel('The Number of Epoch',fontsize=20)
plt.ylabel('Eposide Constraint Violation Distance',fontsize=20)
plt.grid()
```
